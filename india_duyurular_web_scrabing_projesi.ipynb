soru2: soup nesnesinden table etiketinden views-table classından verileri bulunuz ve yazdırınız

```python
konusma_tablosu=soup.find('table',class_='views-table')
konusma_tablosu.text
```

soru3: tbody etiketini bulun içerisinden bütün tr etiketlerini ayıklayın ve satırlar isimli değişkene atayın

```python
satırlar=konusma_tablosu.find('tbody')
```

```python
satırlar.text
```

```python
satırlar.find_all('tr')
```

```python
#satırlar=konusma_tablosu.find('tbody').find_all('tr')
```

```python
satırlar = konusma_tablosu.find('tbody').find_all('tr')
satırlar
```

soru4:satırlar içerisinden bir for döngüsü başlatın ve her bir satırdaki sütünları bulmak için td etiketini bulun

```python
konusma_verileri = []
for satir in satırlar:
    sütunlar = satir.find_all('td')
    if len(sütunlar) >= 3:
        sr_no = sütunlar[0].get_text(strip= True)
        issue_date = sütunlar[1].get_text(strip = True)
        title = sütunlar[2].get_text(strip = True)
        link = sütunlar[2].find('a')['href']
        full_link = 'https://presidentofindia.nic.in/' + link # tam url
        konusma_verileri.append((sr_no, issue_date, title, full_link))

for data in konusma_verileri:
    print(data)
```

soru5: yukarıda oluşturduğumuz konuşma verileri isimli datayı dataframe'e dönüştürün ve csv dosyası olarak kaydedin

```python
import pandas as pd
df=pd.DataFrame(data=konusma_verileri,columns=['seri no','issue date','title', 'link'])
df.head()
```

```python
df.to_csv('Hindistan_Duyurular.csv')
```
